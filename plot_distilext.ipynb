```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns; sns.set()
import pandas as pd
import numpy as np
```

```python
layer_comparison = pd.DataFrame({
    'Number of Layers': [1, 2, 3, 4, 5, 6, 7, 8], 
    'CNN/DM': [18.49, 18.55, 18.69, 18.71, 18.06, 18.92, 18.80, 18.97],
    'XSum': [15.38, 15.40, 16.22, 15.86, 15.98, 16.47, 16.23, 16.35]})
```

```python
layer_comparison
```

```python
x = layer_comparison['Number of Layers']
y1 = layer_comparison['CNN/DM']
y2 = layer_comparison['XSum']

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)

# plt.xlabel('Number of Layers')
# plt.ylabel('Scores')

ax.plot(x, y1, marker='o')
ax.plot(x, y2, marker='o')
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

for _x, _y1, _y2 in zip(x, y1, y2):
    ax.annotate(str(_y1), xy=(_x, _y1), xytext=(_x-0.14, _y1 + 0.07))
    ax.annotate(str(_y2), xy=(_x, _y2), xytext=(_x-0.2, _y2 + 0.1))
    
plt.savefig('test.png', bbox_inches = 'tight', pad_inches = 0)
```

```python
inference = pd.DataFrame({
    'Inference': ['CPU(CNN/DM)', 'CPU(XSum)', 'GPU(CNN/DM)', 'GPU(XSum)'],
    'BERTSUMEXT': [0.321373, 0.242089, 0.026463, 0.020679], 
    'DistilEXT': [0.139050, 0.076501, 0.014117, 0.01335]
})
```

```python
inference
```

```python
labels = inference['Inference']
bert = inference['BERTSUMEXT']
distil = inference['DistilEXT']
```

```python
print(labels)
```

```python
font = {'weight' : 'bold',
        'size'   : 11}

plt.rc('font', **font)


y = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)

rects2 = ax.barh(y + width/2, distil, width, label='DistilExt')
rects1 = ax.barh(y - width/2, bert, width, label='BERTSUMEXT', color='orange')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Time Per Document (sec)', fontsize=14, fontweight='bold')
#ax.set_title('Scores by group and gender')
ax.set_yticks(y)
ax.set_yticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        #print(rect)
        width = rect.get_width()
        #height = rect.get_height()
        ax.annotate('{}'.format(width),
                    #xy=(rect.get_x() + rect.get_width() / 2, height),
                    xy=(width, rect.get_y() + rect.get_height() / 2),
                    xytext=(0+30, -5),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
plt.savefig('inference.png', bbox_inches = 'tight', pad_inches = 0)
```
